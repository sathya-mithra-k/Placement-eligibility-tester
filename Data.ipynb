```python
import pandas as pd
import numpy as np

# Generate data for 600 rows
n = 600

# Define possible values for each column
genders = np.random.choice(['Male', 'Female'], size=n)
cgpa_values = np.round(np.random.uniform(6.0, 10.0, size=n), 1)
skills = []
all_skills = ['Java', 'Python', 'C++', 'JavaScript', 'HTML/CSS', 'Teamwork', 'Communication', 'Problem Solving']
for _ in range(n):
    num_skills = np.random.randint(1, 5)  # Randomly choose 1 to 4 skills
    chosen_skills = np.random.choice(all_skills, size=num_skills, replace=False)
    skills.append(', '.join(chosen_skills))
departments = np.random.choice(['CSBS','CSBS', 'ECE', 'MECH','AI & DS' ,'AI & ML', 'CCE'], size=n)
internship_experience = np.random.randint(0, 3, size=n)
aptitude_scores = np.random.randint(50, 100, size=n)
communication_skills = np.round(np.random.uniform(3.0, 10.0, size=n), 1)

# Create the DataFrame
data = {
    'Gender': genders,
    'CGPA': cgpa_values,
    'Skills': skills,
    'Student_Department': departments,
    'Prior_Internship_Experience(Integer)': internship_experience,
    'Aptitude_Score': aptitude_scores,
    'Communication_Skill_Score(out_of_10)': communication_skills
}

df = pd.DataFrame(data)

# Define hiring probabilities with additional conditions
def determine_hiring_status(row):
    if (row['CGPA'] < 7.0) or (row['Prior_Internship_Experience(Integer)'] == 0 and row['Communication_Skill_Score(out_of_10)'] < 5.0 and row['Aptitude_Score'] < 70):
        return 'No'
    else:
        return np.random.choice(['Yes', 'No'], p=[0.9, 0.1])  # Adjust probability as needed

df['Zoho_hired'] = df.apply(determine_hiring_status, axis=1)
df['Accenture_hired'] = df.apply(determine_hiring_status, axis=1)
df['IBM_hired'] = df.apply(determine_hiring_status, axis=1)
df['Cognizant_hired'] = df.apply(determine_hiring_status, axis=1)

# Create a column indicating if the person knows any coding language
def knows_coding_language(skills_list):
    coding_languages = ['Java', 'Python', 'C++', 'JavaScript']
    for skill in skills_list:
        if skill in coding_languages:
            return 'Yes'
    return 'No'

df['Knows Coding Language'] = df['Skills'].apply(knows_coding_language)

# Display the dataset
print(df)
```

```python
df.to_csv('student_dataset.csv', index=False)
```
